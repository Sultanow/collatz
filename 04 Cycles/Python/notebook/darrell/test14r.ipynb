```
# 
# This program finds a rotation of the floor function parity vector that matches the ceiling function parity vector
#
import math
l=211
for n in range(1,l):
   sv=[0 for i in range(10000)]
   pv=[0 for i in range(10000)]
   r=[0 for i in range(10000)]
   ws=0
   wp=0
   for j in range(1,l+1):
      a=j*n
      a=math.ceil(a/l)
      b=(j-1)*n
      b=math.ceil(b/l)
      c=j*n
      c=math.floor(c/l)
      d=(j-1)*n
      d=math.floor(d/l)
      sv[j-1]=c-d
      pv[j-1]=a-b
      ws=ws*2+c-d
      wp=wp*2+a-b
  #    print(sv[j-1],pv[j-1])
  # print("floor ceiling",ws,wp)
   mask=(2<<(l-1))-1
   a=ws
   r[0]=a
   for j in range(1,l+1):
      b=a>>(l-1)
      if (b==0):
         a=(a*2)&mask
      else:
         a=((a*2)&mask)+1
      r[j]=a
  #    print(r[j])
   for j in range(1,l+1):
      if (r[j]==wp):
         print("n",n,"match",j,"basis",math.floor((j*n+1)/l))
         break



       



```
