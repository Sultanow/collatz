```
#
# This program computes R(N-x)-(L-r)x=L-1
# Darrell Cox, 12/22/20
#
# For prime N, x and N-x are relatively prime.
# When N is a power of 2, the x and N-x values are divisible by 4 (but neither by 8)
#   
import math
l=215
out=2
sc=[0 for i in range(10000)]
sf=[0 for i in range(10000)]
r=[0 for i in range (10000)]
mask=(2<<(l-1))-1
for i in range(1,l):
   n=i
   sumc=0
   sumf=0
   pc=0
   pf=0
   for j in range(1,l+1):
      a=j*n
      a=math.ceil(a/l)
      b=(j-1)*n
      b=math.ceil(b/l)
      sumc=sumc+a-b
      sc[j-1]=sumc
      pc=pc*2+a-b
      a=j*n
      a=math.floor(a/l)
      b=(j-1)*n
      b=math.floor(b/l)
      sumf=sumf+a-b
      sf[j-1]=sumf
      pf=pf*2+a-b
   pf=pf*2
 #  print(pc,pf)
   a=pf
   r[0]=a
   for j in range(1,l):
      b=a>>(l-1)
      if (b==0):
         a=(a*2)&(mask)
      else:
         a=((a*2)&mask)+1
      r[j]=a
 # print(r[0],r[1],r[2],r[3],r[4],r[5],r[6])
   rot=0
   for j in range(1,l):
      if (r[j-1]==pc):
         rot=j
 #       print("rotation=",rot)
         break
   for j in range(1,rot+1):
      temp=sf[0]
      for k in range(1,l):
         sf[k-1]=sf[k]
      sf[l-1]=temp
   if (out==0):
      for j in range(1,l+1):
         print(sf[j-1],sc[j-1],sf[j-1]-sc[j-1])
      print("\n")
   else:
      if (out==1):
         print(i," ",sf[0],sc[0],sf[0]-sc[0]," ",sf[l-1],sc[l-1],sf[l-1]-sc[l-1]," ",rot)
      else:
         print(i," ",sf[0]-sc[0],sc[l-1]-sf[l-1]," ",rot)




```
